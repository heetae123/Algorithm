# 필수 6.8

## 1) 문제 정의 (MD)

알파벳으로 {알,고,리,즘}을 가질 수 있는 다음 리스트를 카운팅 정렬을 이용해 사전순서로 정렬해 보자.

## 2) 문제 정의 (MD)

카운팅 정렬을 이용하자.

## 3) 손으로 푼 예제 (MD, 이미지 삽입)

##  4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)


리스트를 입력 변수로 받아 출력 변수는 정렬된 리스트가 나온다.
함수 자체 코드는 임시 리스트를 만들어 각 숫자의 빈도를 계산 및 count라는 변수에 저장하고 임시 리스트에서 원소를 꺼내어 쓸 때마다 빈도 수의 -1을 해준다 이렇게 정렬 결과를 원래 배열에 복사를 하게 된다

##  5) 코드 (PY,C)

```python
def counting_sort(A):
   
    MAX_VAL = ord(max(A)) + 1
    output = [''] * len(A)
    count = [0] * MAX_VAL

   
    for char in A:
        count[ord(char)] += 1


    for i in range(1, MAX_VAL):
        count[i] += count[i - 1]

  
    for char in reversed(A):
        output[count[ord(char)] - 1] = char
        count[ord(char)] -= 1

   
    for i in range(len(A)):
        A[i] = output[i]
```

##  6) 테스트 코드  (PY,C)

## 2) 문제 정의 (MD)

## 3) 손으로 푼 예제 (MD, 이미지 삽입)

##  4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

리스트를 입력 변수로 받아 출력 변수는 정렬된 리스트가 나온다.
함수 자체 코드는 임시 리스트를 만들어 각 숫자의 빈도를 계산 및 count라는 변수에 저장하고 임시 리스트에서 원소를 꺼내어 쓸 때마다 빈도 수의 -1을 해준다 이렇게 정렬 결과를 원래 배열에 복사를 하게 된다

##  5) 코드 (PY,C)

```python
def counting_sort(A):
    
    RANGE = 52  
    output = [''] * len(A)
    count = [0] * RANGE

    # Map characters to indices (a-z -> 0-25, A-Z -> 26-51)
    def char_to_index(char):
        if 'a' <= char <= 'z':
            return ord(char) - ord('a')
        else:
            return ord(char) - ord('A') + 26

 
    def index_to_char(index):
        if 0 <= index < 26:
            return chr(index + ord('a'))
        else:
            return chr(index - 26 + ord('A'))

   
    for char in A:
        count[char_to_index(char)] += 1

   
    for i in range(1, RANGE):
        count[i] += count[i - 1]

   
    for char in reversed(A):
        index = char_to_index(char)
        output[count[index] - 1] = char
        count[index] -= 1

 
    for i in range(len(A)):
        A[i] = output[i]
```
