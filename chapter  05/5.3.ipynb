## 2) 알고리즘 설명 

배열에서 하나의 요소를 선택해 이를 기준으로 삼는다.
하나의 요소를 선택한 기준 왼쪽은 그보다 작은 값을 오른쪽은 그보다 큰 값을 위치한다.
또한 하나의 요소를 선택한 기준보다 작은 값들의 하나의 요소를 또 선택해 이를 기준으로 잡고 그보다 작은 것은 왼쪽 그보다 큰 것은
오른쪽으로 놔 계속해 반복한다

## 4) 코드 개요

리스트 A와 리스트의 정렬 범위를 나타내는 left와 right를 입력으로 받는다.
정렬 범위가 2개 이상인 경우 좌우로 분할하고 왼쪽 부분부터 정렬을 끝내고 끝나면 오른쪽 정렬을 한다

## 5) 코드

```python
def quick_sort(A,left,right):
    if left<right :
        mid=partition(A,left,right)
        quick_sort(A,left,mid-1)
        quick_sort(A,mid+1,right)
```

## 6) 테스트 코드

```python
def partition(A, left, right):
    low = left + 1
    high = right
    pivot = A[left]
    while low <= high:
        while low <= right and A[low] <= pivot:
            low += 1
        while high >= left and A[high] > pivot:
            high -= 1

        if low < high:
            A[low], A[high] = A[high], A[low]

    A[left], A[high] = A[high], A[left]
    return high

def quick_sort(A, left, right):
    if left < right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid - 1)
        quick_sort(A, mid + 1, right)

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
quick_sort(data, 0, len(data) - 1)
print("QuickSort: ", data)
```
