이 트리를 전위 순회한 결과를 적어라

6-4-2-1-3-5-9-7-10-8-11

(루트 노드를 방문한 다음에 왼쪽부터 오른쪽 방문)

이 트리를 중위 순회한 결과를 적어라

1-2-3-4-5-6-7-9-8-10-11

(왼쪽 노드를 먼저 방문한 다음에 루트 방문 -> 오른쪽 방문)

이 트리를 후위 순회한 결과를 적어라

1-3-2-5-4-7-8-11-9-10-6

(왼쪽과 오른쪽 방문 후 루트를 방문)

### 코드 1) -> 전위 순회

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def preorder(n):
    if n is not None:
        print(n.data, end='-')
        preorder(n.left)
        preorder(n.right)


node1 = TNode('1')
node3 = TNode('3')
node2 = TNode('2', node1, node3) 
node5 = TNode('5')
node4 = TNode('4', node2, node5)  
node7 = TNode('7')
node8 = TNode('8')
node11 = TNode('11')
node10 = TNode('10', node8, node11)  
node9 = TNode('9', node7, node10)   
root = TNode('6', node4, node9)     


preorder(root)
```

### 코드 2) -> 중위 순회

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def preorder(n):
    if n is not None:
        preorder(n.left)
        print(n.data, end='-')
        preorder(n.right)


node1 = TNode('1')
node3 = TNode('3')
node2 = TNode('2', node1, node3) 
node5 = TNode('5')
node4 = TNode('4', node2, node5)  
node7 = TNode('7')
node8 = TNode('8')
node11 = TNode('11')
node10 = TNode('10', node8, node11)  
node9 = TNode('9', node7, node10)   
root = TNode('6', node4, node9)     


preorder(root)
```

### 코드 3) -> 후위 순회

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def preorder(n):
    if n is not None:
        preorder(n.left)
        preorder(n.right)
        print(n.data, end='-')

node1 = TNode('1')
node3 = TNode('3')
node2 = TNode('2', node1, node3) 
node5 = TNode('5')
node4 = TNode('4', node2, node5)  
node7 = TNode('7')
node8 = TNode('8')
node11 = TNode('11')
node10 = TNode('10', node8, node11)  
node9 = TNode('9', node7, node10)   
root = TNode('6', node4, node9)     


preorder(root)
```
