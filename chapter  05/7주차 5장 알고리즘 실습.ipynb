# 분할 정복 기법

## 문제 정의

병합 정렬

## 알고리즘 설명

부분 리스트의 사이즈가 1이 될 때까지 크기가 같은 두개의 부분 리스트로 분할을 반복한다
크기가 1인 분할된 부분리스트들을 합치면서 정렬을 시작한다.

## 코드 개요

## 알고리즘 코드
