# 5.4

##   1) 문제 정의 

이진 트리 높이

## 2) 알고리즘 설명 

이진 트리의 왼쪽 높이와 오른쪽 높이를 구할 수 있다고 가정한다면 둘 중 더 큰 높이를 비교해서 찾고 그 높이의 root 노드를 
더하면 된다

## 3) 손으로 푼 예제

## 4) 코드 개요

입력변수는 루트 노드로 어떤 트리의 루트를 나타낸다 
출력 변수는 루트 노드 포함한 노드의 레벨들을 출력한다.
(공백 트리가 될 때까지 두 서브 트리의 높이를 계산하는데 만약 공백트리가 나오면 그 즉시 중단한다)

## 5) 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
```

## 6) 테스트 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
        
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('f', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)
print("트리의 높이=", calc_height(root))
```

##  8) 복잡도 분석

T(n)=T(left)+T(right)+1 여기서 +1이 추가된 이유는 마지막 공백트리가 아니라면 left, right를 비교하고 더 큰 함수+1을 하는 연산이 있기에 쓴 것이다
하지만 이 시간 복잡도는 내부 노드를 통해서 계산한 값이다 만약 외부 노드인 공백트리까지 계산한다면 T(n)=2n+1이 된다(외부 노드는 자식이 없는 공백트리이며, 모든 내부 노드는 두개의 자식 트리를 갖기 때문에 2n이며 루트노드를 더해야 하므로 2n+1이 된다)
하지만 둘 다 O(N)으로 같다.

# 5.5~5.7

##   1) 문제 정의 

이진트리의 표준 순회

## 2) 알고리즘 설명 

표준 순회는 전위 순회, 중위 순회, 후위 순회가 있다. 
전위 순회는 루트 노드를 방문한 다음에 왼쪽부터 오른쪽으로 방문하게 되는 것이며
중위 순회는 왼쪽 노드를 먼저 방문한 다음에 루트 노드를 방문하고 마지막으로 오른쪽을 방문한다.
후위 순회는 왼쪽과 오른쪽 노드를 방문 후 루트 노드를 마지막으로 방문한다.

## 3) 손으로 푼 예제

## 4) 코드 개요

루트 노드를 입력 변수로 지정하고 전위 순회 함수의 경우 root를 먼저 처리하고 left의 공백 트리가 있을 때까지 순환하며 끝이 나면 right의 공백트리가 있을 때까지 처리
//
중위 순회 함수의 경우 left의 공백 트리가 있을 때까지 순환하며 끝이 나면 root 노드를 처리하고 그 후 right의 공백트리가 있을 때까지 처리
//
후위 순회 함수의 경우 left의 공백 트리가 있을 때까지 순환하며 끝이 나면 right의 공백트리가 있을 때까지 처리 마지막으로 root 노드 처리

## 5) 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right
// 전위 순회
def preorder(n) :
    if n is not None :
        print(n.data, end='  ')
        preorder(n.left)
        preorder(n.right)
// 중위 순회
def inorder(n):
    if n is not None:
        inorder(n.left)
        print(n.data, end='  ')
        inorder(n.right)
//후위 순회
def postorder(n):
    if n is not None:
        postorder(n.left)
        postorder(n.right)
        print(n.data, end='  ')
        
```

## 6) 테스트 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right
def preorder(n) :
    if n is not None :
        print(n.data, end='  ')
        preorder(n.left)
        preorder(n.right)
def inorder(n):
    if n is not None:
        inorder(n.left)
        print(n.data, end='  ')
        inorder(n.right)

def postorder(n):
    if n is not None:
        postorder(n.left)
        postorder(n.right)
        print(n.data, end='  ')
        
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('f', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)
print('  In-order : ', end='')
inorder(root)
print()
print('  Pre_Order : ', end='')
preorder(root)
print()
print('  Post_Order : ', end='')
postorder(root)
print()
```
