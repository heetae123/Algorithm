### 8.1
거스름돈 동전 최소화

### 1) 문제 정의
액면가가 서로 다른 m가지의 동전이 있다.
거스름돈으로 v원을 동전으로만 돌려주어야 한다면 최소 몇 개의 동전이 필요한지를 구하라.
단, 모든 동전은 무한히 사용할 수 있고, 동전들은 액수가 큰 것부터 내림차순으로 순서대로 정렬되어 있다.

### 2) 알고리즘 설명 
액면가가 가장 큰 동전을 사용하고 남은 잔돈들도 마찬가지로 큰 동전으로 사용하며, 최소한의 동전 수로 잔돈에 맞게 설정한다

### 3) 손으로 푼 예제

###   4) 코드 개요
어느 종류의 동전과 내야하는 값을 인자로 받고, 어떤 종류의 코인 길이만큼 반복문을 돌려 만약 coins와 내야 하는 금액을 나눠서 몫이 정수가 나오면 count 리스트에 저장 그 후 전체 금액에서 정수가 나온 coin의 종류를 빼서 리턴함

###  5) 코드

```python

def min_coins_greedy( coins, V ) :
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count
```

### 6) 테스트 코드

```python

def min_coins_greedy( coins, V ) :
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count
coins =[1000,500,100,50,10]
v=2260
print("잔돈 액수=",v)
print("동전 종류= ",coins)
print("동전 개수 = ",min_coins_greedy(coins,v))
```

### 8.2

### 1) 문제 정의
각각 무게가 w이고 가치가 v인 n개의 물건들이 있고, 이것을 ㅂ ㅐ낭에 넣으려고 한다.
이때, 배낭에 넣을 수 있는 용량은 w를 초과하지 않아야 하고, 물건들의 가치의 합이 최대가 되도록 배낭을 채우고, 이때 배낭의 최대가치를 구하라.
단, 물건들은 분할해서 일부분만을 넣을 수도 있다.

### 2) 알고리즘 설명 
0-1 배낭 채우기의 심화 과정이다.
0-1배낭 채우기의 알고리즘은 분할해서 채우진 않았지만 이 알고리즘 같은 경우 물건들의 무게를 최대치로 채우고 나서 남은 배낭의 무게를 보고 두번째 가치가 높은 물건을 다 채우는 것이 아닌 분할해서 채우는 알고리즘이다

### 3) 손으로 푼 예제


###   4) 코드 개요
물건들의 종류와 그의 따른 무게, 가치와 총 무게를 입력받아 가치가 제일 높은 물건들을 먼저 채우고 그 다음의 가치가 높은 물건들을 채우는 알고리즘이다.
따라서, 용량이 다 찬 경우는 반복을 멈추고 만약 용량이 다 차지 않는 경우 물건들의 종류의 갯수만큼 반복문을 돌린다

### 5) 코드

```python
def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)
    
    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W -= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))
            
    return totalValue
```

### 6) 테스트 코드

```python
def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)
    
    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W -= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))
            
    return totalValue
obj=[("A",10,80),("B",12,120),("C",7,60)]
print("w=18",obj)
print("부분적인 배낭(18):",knapSack_fractional_greedy(obj,18),end="\n\n")
```

### 6) 테스트 코드

```python
def getMinVertex(dist, selected):
    minDist = float('inf')
    minVertex = -1
    for v in range(len(dist)):
        if not selected[v] and dist[v] < minDist:
            minDist = dist[v]
            minVertex = v
    return minVertex

def MSTPrim(vertex, adj):
    vsize = len(vertex)
    dist = [float('inf')] * vsize
    dist[0] = 0
    selected = [False] * vsize
    parent = [-1] * vsize

    for i in range(vsize):
        u = getMinVertex(dist, selected)
        if u == -1:
            break
        selected[u] = True
        print(vertex[u], end=':')
        print(dist)

        for v in range(vsize):
            if adj[u][v] is not None:
                if not selected[v] and adj[u][v] < dist[v]:
                    dist[v] = adj[u][v]
                    parent[v] = u

    print("Edges in MST:")
    for i in range(1, vsize):
        if parent[i] != -1:
            print(f"{vertex[parent[i]]} - {vertex[i]}, weight: {adj[parent[i]][i]}")

vertex = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
weight = [
    [None, 29, None, None, None, 10, None],
    [29, None, 16, None, None, None, 15],
    [None, 16, None, 12, None, None, None],
    [None, None, 12, None, 22, None, 18],
    [None, None, None, 22, None, 27, 25],
    [10, None, None, None, 27, None, None],
    [None, 15, None, 18, 25, None, None]
]

print("MST By Prim's Algorithm")
MSTPrim(vertex, weight)
```

### 8.8

### 1) 문제 정의
집단 s에 포함되지 않은 정점들 중에서 집합 s와 가장 가까운 정점을 선택한다.
즉, 최단거리가 확정된 정점들과 간선으로 직접 연결된 정점들 중에서 가장 가까운 정점을 선택한다.

### 2) 알고리즘 설명 
u,v,w 중의 dist값이 최소라면 한번에 가는 방법보다 시작정점을 u,v,w에 거쳐서 최단 거리를 찾는 알고리즘이다

### 3) 손으로 푼 예제



###   4) 코드 개요
정점 리스트와 인접 행렬, 시작 정점의 인덱스를 입력하고, 시작정점의 거리는 0으로 초기화 시켜놓고 그와 인접한 정점들과의 거리가 최소인 부분을 찾는다.
즉, 시작 정점에서 어느 정점을 거쳐서 마지막 정점을 가는 것보다 바로 가는 것이 만약에 최단 거린 지 아닌지 검사하여 만약 최단 거리라면 이 부분을 선택하게 끔 한다

### 5) 코드

```python
def shortest_path_dijkstra(vtx, adj, start) :
    vsize = len(vtx)
    dist = list(adj[start])
    dist[start] = 0
    path = [start] * vsize
    found = [False] * vsize
    found[start] = True
    
    for i in range(vsize) :
        print("Step%2d: "%(i+1),dist)
        u = getMinVertex(dist, found)
        found[u] = True
        
        for w in range(vsize) :
            if not found[w] :
                if dist[u] + adj[u][w] < dist[w] :
                    dist[w] = dist[u] + adj[u][w]
                    path[w] = u
                    
    return path
```

### 6) 테스트 코드

```python
import sys

def getMinVertex(dist, found):
    min_dist = sys.maxsize
    min_vertex = -1
    
    for v in range(len(dist)):
        if not found[v] and dist[v] < min_dist:
            min_dist = dist[v]
            min_vertex = v
    return min_vertex

def shortest_path_dijkstra(vtx, adj, start):
    vsize = len(vtx)
    dist = list(adj[start])
    dist[start] = 0
    path = [start] * vsize
    found = [False] * vsize
    found[start] = True
    
    for i in range(vsize - 1):  
        print("Step%2d: " % (i+1), dist)
        u = getMinVertex(dist, found)
        found[u] = True
        
        for w in range(vsize):
            if not found[w]:
                if dist[u] + adj[u][w] < dist[w]:
                    dist[w] = dist[u] + adj[u][w]
                    path[w] = u

    return path, dist

vtx = ['A', 'B', 'C', 'D', 'E']
adj = [
    [0, 10, 3, sys.maxsize, sys.maxsize],
    [10, 0, 1, 2, sys.maxsize],
    [3, 1, 0, 8, 2],
    [sys.maxsize, 2, 8, 0, 7],
    [sys.maxsize, sys.maxsize, 2, 7, 0]
]

start = 0  

path, dist = shortest_path_dijkstra(vtx, adj, start)
print("최단 경로: ", path)
print("최단 거리: ", dist)
```
