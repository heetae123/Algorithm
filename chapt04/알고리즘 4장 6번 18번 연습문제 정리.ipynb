셸 정렬은 버블 정렬과 비슷하지만 조사 간격을 처음부터 큰 수를 점차 작은 수로 줄여 효율적으로 정렬하는 것이다.
따라서 삽입 정렬은 어떠한 숫자를 일단 뽑아서 그 숫자가 들어갈 위치를 찾는 것이지만
셸 정렬은 버블 정렬과 비슷한 방법으로 억지기법처럼 일일이 다 비교해본다 하지만 그 조사 간격이 버블 정렬과 달리 점차적으로 줄어들게 만들어서 효율적으로 비교할 수 있도록 하는 것이 차이점이다

```python
def partition(a, left, right):
    pivot_index = left
    pivot_value = a[pivot_index]

    while left < right:
        while left < len(a) and a[left] <= pivot_value:
            left += 1
        while a[right] > pivot_value:
            right -= 1
        if left < right:
            a[left], a[right] = a[right], a[left]

    a[right], a[pivot_index] = a[pivot_index], a[right]
    return right


def quick_select_median(a):
    left = 0
    right = len(a) - 1
    target_index = len(a) // 2

    while True:
        pivot_index = partition(a, left, right)

        if pivot_index == target_index:
            return a[pivot_index]
        elif pivot_index < target_index:
            left = pivot_index + 1
        else:
            right = pivot_index - 1


a = [12, 5, 7, 9, 18, 3, 8]
median = quick_select_median(a)
print("Median:", median)
```

```python
def partition(a, left, right):
    pivot_index = left
    pivot_value = a[pivot_index]

    while left < right:
        while left < len(a) and a[left] <= pivot_value:
            left += 1
        while a[right] > pivot_value:
            right -= 1
        if left < right:
            a[left], a[right] = a[right], a[left]

    a[right], a[pivot_index] = a[pivot_index], a[right]
    return right
```

while문을 돌린다,
while은 left와 right가 극한으로 가면 갈 수록 둘이 역전되는 상황이 되는데 그 상황이 나올 경우 종료하겠다라는 뜻이다
이렇게 역전되지 않을 정도로 반복을 하는데 left값이 피벗보다 작다고 하면 패스하면서 left 인덱스 값을 증가시킨다 만약 피벗보다 큰 값이 나오면 잠깐 스탑
그다음 right가 피벗보다 값이 크면 패스 아니라면 잠깐 스탑한다 
이렇게 된 상황에서 아직 left와 right가 역전 되지 않는 경우라면 둘의 값을 스왑한다라는 뜻이다

```python
def quick_select_median(a):
    left = 0
    right = len(a) - 1
    target_index = len(a) // 2

    while True:
        pivot_index = partition(a, left, right)

        if pivot_index == target_index:
            return a[pivot_index]
        elif pivot_index < target_index:
            left = pivot_index + 1
        else:
            right = pivot_index - 1
```
