## 알고리즘 설명

반복 구조를 통해 한 변수에 반복적으로 계산하여 반환한다.

## 알고리즘 코드

```python
def factor(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result
```

## 테스트 코드

```python
def factor(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result
num=int(input("정수를 입력하시오"))
number=factor(num)
print(number)
```

## 2)문제 정의(하향식)

순환 구조로 사용

## 알고리즘 개요

순환 구조를 통해 문제의 크기를 (n-1)로 줄어 들게 한다.

## 알고리즘 설명

순환 구조를 통해 한 변수에 반복적으로 계산하여 반환한다.

## 알고리즘 코드

```python
def factor(n):
    if n==1:
        return 1
    else:
        return(n*factor(n-1))
```

## 테스트 코드

```python
def factorial(n):
    if n < 0:
        return "음수의 팩토리얼은 정의되지 않습니다."
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num=int(input("정수를 입력하시오"))
number=factorial(num)
print(number)
```

## 시간복잡도

O(n!)

# 문제 정의

방향 그래프 G=(V,E)가 주어졌다.
G에 존재하는 각 정점들의 선행 순서를 위배하지 않으면서 모든 정점들을 순서대로 나열하라.

# 알고리즘 개요 

방향 그래프에서의 모든 정점을 한 번씩 방문한다

# 알고리즘 설명

방향 그래프에서 진입 차수가 0인 정점 중에서 하나를 임의로 선택하고
선택된 정점은 삭제한다.
정점이 삭제되므로 이 정점에서 진출하는 모든 간선들도 삭제한다.
이때, 간선이 삭제되므로 삭제되는 간선과 연결된 이웃 정점들의 진입 차수는 갱신되어야 한다.
이 과정을 모든 정점이 삭제될 때까지 반복한다.

# 알고리즘 코드

```python
def topological_sort(graph):
    inDeg={}
    for v in graph:
        inDeg[v]=0
    for v in graph:
        for u in graph[v]:
            inDeg[u]+=1
    vlist=[]
    for v in graph:
        if inDeg[v]==0:
            vlist.append(v)
    while vlist:
        v=vlist.pop()
        print(v, end='')
        for u in graph[v]:
            inDeg[u]-=1
            if inDeg[u]==0:
                vlist.append(u)
```

```python
def topological_sort(graph):
    inDeg={}
    for v in graph:
        inDeg[v]=0
    for v in graph:
        for u in graph[v]:
            inDeg[u]+=1
    vlist=[]
    for v in graph:
        if inDeg[v]==0:
            vlist.append(v)
    while vlist:
        v=vlist.pop()
        print(v, end='')
        for u in graph[v]:
            inDeg[u]-=1
            if inDeg[u]==0:
                vlist.append(u)
mygraph={"A" : {"C","D"},
         "C" : {"D","F"},
         "B" : {"D","E"},
         "D" : {"F"},
         "E" : {"F"},
         "F" : {}}
print('topological_sort : ')
topological_sort(mygraph)
print()
```

# 5)이진 탐색(순환 구조, 반복 구조)

# 문제 정의

리스트에 n개의 항목이 들어있다.
이 리스트에서 "탐색키"를 가진 항목을 찾아라.
단, 리스트의 항목들은 정렬되어 있다.

# 알고리즘 개요 

축소 정복 기법의 한 종류로써,
문제의 크기가 절반으로 줄어든다.(리스트의 중앙에 있는 항목을 먼저 조사한다.
만약 이 항목이 탐색키보다 크다면 찾는 항목은 왼쪽에 있고, 오른쪽은 더 이상 검사할 필요가 없다.
반대로 그 항목이 탐색키보다 작다면 답은 오른쪽에 있고, 왼쪽은 검사할 필요가 없다.
따라서 각 단계마다 검사할 범위, 즉 문제의 크기가 반으로 줄어든다)

# 알고리즘 설명

정렬된 리스트 A와 key값, 그리고 A의 탐색 범위인 low와 high가 제공되어야 한다. 탐색에 성공하면 항목의 인덱스를, 실패하면 -1을 반환한다

# 알고리즘 코드(순환 구조)

```python
def search(A,key,low,high):
    if(low<=high):
        mid=(low+high)//2
        if key==A[mid]:
            return mid
        elif key<A[mid]:
            return search(A,key,low,mid-1)
        else:
            return search(A,key,mid+1,high)
        return -1
```

# 알고리즘 코드(반복 구조)(실행 효율성을 위해서는 이 코드가 더 유리)

```python
def search(A,key,low,high):
    if(low<=high):
        mid=(low+high)//2
        if key==A[mid]:
            return mid
        elif key>A[mid]:
            low=mid+1
        else:
            high = mid-1
        return -1
```

# 테스트 코드

```python
def search(A, key, low, high):
    if low <= high:
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key < A[mid]:
            return search(A, key, low, mid - 1)
        else:
            return search(A, key, mid + 1, high)
    return -1

listA = [1, 3, 8, 13, 13, 16, 21, 26, 27, 30, 33, 36, 39, 41, 44, 49]
print("입력 리스트 =", listA)
print("33 탐색:", search(listA, 33, 0, len(listA) - 1))
```

# 실행결과

# 시간 복잡도

최선의 경우 : O(1)
최악의 경우 : O(logN)
