### 1) 문제 정의

메모이제이션 동적계획과 테이블화 동적계획법

### 2) 알고리즘 설명

메모이제이션 동적 계획법은 한번 구한 것을 여러번 다시 구하지 않고 새로운 리스트에 저장하여 때마다 사용하게끔 만드는 기법
테이블화 동적 계획법은 기반 상황을 먼저 계산하여 테이블을 만들고 그 이후 상황들은 기반 상황을 이용해 계산하는 방법

### 3) 손으로 푼 예제

이 순환 식을 알면 메모이제이션과 테이블화를 만들 수 있다.

### 4) 알고리즘 개요

메모이제이션 : 새로운 메모리를 한개 먼저 만들어 놓고 None으로 초기화시킨다 이 의미는 어떤 문제도 풀지 못했다는 의미이며,
그 이후 n<2인 경우, 0과 1은 피보나치 수열에서 0과 1을 의미하므로 고정시켜놓고 n>=2이상인 경우, n-1과 n-2의 값을 더해 리턴한다

테이블화 : 새로운 테이블을 None으로 초기화시켜주고 기반 상황에 대한 테이블 항목을 채운다.
그 이후 부분 문제들을 기반 상황의 값으로 계산하여 문제를 풀이한다.

### 5) 알고리즘 코드

```python
def fib(n) :
    if (mem[n] == None):
        if n < 2:
            mem[n] = n
        else:
            mem[n] = fib(n-1) + fib(n-2)
    return mem[n]
def fib_dp_tab(n):
    f=[None]*(n+1)
    f[0]=0
    f[1]=1
    for i in range(2,n+1):
        f[i]=f[i-1]+f[i-2]
    return f[n]
n = 8
mem = [None] * (n+1)
print("메모이 제이션 결과값: ", fib(n))
print("테이블 화 결과값: ", fib_dp_tab(n)
```

### 6) 테스트 코드

```python
def fib(n) :
    if (mem[n] == None):
        if n < 2:
            mem[n] = n
        else:
            mem[n] = fib(n-1) + fib(n-2)
    return mem[n]
def fib_dp_tab(n):
    f=[None]*(n+1)
    f[0]=0
    f[1]=1
    for i in range(2,n+1):
        f[i]=f[i-1]+f[i-2]
    return f[n]
```
